## Imports

```
import tensorflow as tf
import numpy as np
from tensorflow.keras.datasets import imdb
```

## Preprocessing

### Load dataset

Original dataset https://www.kaggle.com/lakshmi25npathi/imdb-dataset-of-50k-movie-reviews

```
number_of_words = 20000
max_len = 100
```

```
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=number_of_words)
```

### Padding

```
X_train = tf.keras.preprocessing.sequence.pad_sequences(X_train, maxlen=max_len)
X_test = tf.keras.preprocessing.sequence.pad_sequences(X_test, maxlen=max_len)
```

```
print(len(X_train[0]))
print(len(X_train[1]))
```

```
X_train.shape
```

## Learning

### Building CNN

```
X_input = tf.keras.layers.Input((100))

X = tf.keras.layers.Embedding(input_dim=number_of_words, output_dim=128)(X_input)
X = tf.keras.layers.LSTM(units=128, activation='tanh')(X) #return_sequences=True
X = tf.keras.layers.Dense(units=1, activation='sigmoid')(X)

model = tf.keras.models.Model(inputs=X_input, outputs=X)

model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])

model.summary()
```

### Train

```
model.fit(X_train, y_train, epochs=3, batch_size=128)
```

## Evaluation

```
test_loss, test_acurracy = model.evaluate(X_test, y_test)
```
